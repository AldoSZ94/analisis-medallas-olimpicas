### **An치lisis de medallas ol칤mpicas**

#### **Objetivo**

Realizar un an치lisis exploratorio de datos (EDA) sobre un conjunto de datos de medallas ol칤mpicas utilizando **Pandas**. Este proyecto demuestra los conceptos aprendidos sobre **Series**, **DataFrames**, **limpieza de datos**, **filtrado** y **agregaci칩n** en **Pandas**.

#### **Consigna**

Trabajar con el conjunto de datos _medallas.csv_ que incluye informaci칩n sobre las medallas de oro, plata, bronce y el total obtenido por cada pa칤s en los Juegos Ol칤mpicos, para realizar una serie de tareas b치sicas que permitan responder a las preguntas del **cuestionario**.

Las tareas que se realizar치n son las siguientes:

1. **Cargar los Datos**: Importar los datos desde el archivo CSV a un DataFrame de Pandas.
2. **Exploraci칩n inicial**: Utilizar m칠todos b치sicos para explorar el tama침o, las columnas y los tipos de datos del DataFrame.
3. **Limpieza de Datos**: Identificar y manejar valores faltantes o incorrectos, especialmente en las columnas de medallas donde los valores faltantes indican cero medallas.
4. **An치lisis de medallas de oro por pa칤s**: Realizar operaciones que sean necesarias para identificar cu치les fueron los 3 pa칤ses con m치s medallas de oro en total.
5. **An치lisis de medallas totales por pa칤s**: Obtener un DataFrame que contenga s칩lo los pa칤ses que ganaron m치s de 10 medallas en total.


```python
import pandas as pd

# 游댳 Carga el archivo CSV.
df = pd.read_csv("medallas.csv")

# 游댳 Exploraci칩n b치sica del DataFrame.
print(df.shape)
print(df.columns)
df.info()

# 游댳 Estad칤sticas descriptivas.
print(df.describe())

# 游댳 Normaliza nombres de columnas a min칰sculas.
df.columns = df.columns.str.lower()

# 游댳 Reordena columnas.
columnas_ordenadas = ["pais", "oro", "plata", "bronce", "total"]
df = df[columnas_ordenadas]

# 游댳 Muestra primeras y 칰ltimas filas.
print(df.head())
print(df.tail())

# 游댳 Cuenta valores nulos en columnas num칠ricas.
print(df.isnull().sum())

# 游댳 Sustituye nulos por 0 y convierte a enteros.
df[["oro", "plata", "bronce"]] = df[["oro", "plata", "bronce"]].fillna(0).astype("int64")

# 游댳 Verificar si hay pa칤ses repetidos.
print(df["pais"].duplicated().sum())

# 游댳 Agrupa y ordena pa칤ses por medallas de oro.
oro_por_pais = df.groupby("pais")["oro"].sum().sort_values(ascending=False)

# 游댳 Agrupa y ordena pa칤ses por total de medallas.
medallas_por_pais = df.groupby("pais")["total"].sum().sort_values(ascending=False)

# 游댳 Muestra el 2do. pa칤s con m치s medallas de oro.
print(f"游볟 Segundo pa칤s con m치s medallas de oro: {oro_por_pais.index[1]}, con {oro_por_pais.iloc[1]} medallas de oro.")

# 游댳 Muestra el pa칤s con m치s medallas en total.
print(f"游볞 Pa칤s con m치s medallas: {medallas_por_pais.index[0]}, con {medallas_por_pais.iloc[0]} medallas en total.")
```
